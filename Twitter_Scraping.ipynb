```python
import pandas as pd
from tqdm.notebook import tqdm
import snscrape.modules.twitter as sntwitter
```

```python
maxTweets = 100
tweets_list = []

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:@cristiano').get_items()):
    
    data = [
        tweet.date,
        tweet.id,
        tweet.rawContent,
        tweet.user.username,
        tweet.lang,
        tweet.viewCount,
        tweet.replyCount,
        tweet.likeCount,
        tweet.retweetCount,
    ]
    
    tweets_list.append(data)
    if i>maxTweets:
        break
   
```

```python
tweets_list
```

```python
tweet_data = pd.DataFrame(tweets_list, columns = [
                                            'Date Time',
                                            'Tweet id',
                                            'Tweet Content',
                                            'Username',
                                            'Tweet Language',
                                            'Tweet Views',
                                            'Reply Count',
                                            'like Count',
                                            'Retweet Count',
                                        ]
                          )

tweet_data.head()
tweet_data.to_csv("Ronaldo Tweets.csv",index=False)
```
